```python



# Calculate quantiles and IQR for Score
Q1_score = filtered_asian_df['score'].quantile(0.25)
Q3_score = filtered_asian_df['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

# Calculate quantiles and IQR for GDP per Capita
Q1_gdp = filtered_asian_df['Log GDP per capita'].quantile(0.25)
Q3_gdp = filtered_asian_df['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Define outlier bounds
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score
lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# Identify outliers
outliers = filtered_asian_df[
    (filtered_asian_df['score'] < lower_bound_score) | (filtered_asian_df['score'] > upper_bound_score) |
    (filtered_asian_df['Log GDP per capita'] < lower_bound_gdp) | (filtered_asian_df['Log GDP per capita'] > upper_bound_gdp)
]

print("Outlier Countries:")
print(outliers[['Country name', 'score', 'Log GDP per capita']])
```

```python

def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

# South Asia
south_asia_outliers_score = find_outliers_iqr(filtered_asian_df, 'score')
south_asia_outliers_gdp = find_outliers_iqr(filtered_asian_df, 'Log GDP per capita')

# Middle East
middle_east_outliers_score = find_outliers_iqr(filtered_middle_east_df, 'score')
middle_east_outliers_gdp = find_outliers_iqr(filtered_middle_east_df, 'Log GDP per capita')

# Combine outliers from both metrics for each region
south_asia_outliers = pd.concat([south_asia_outliers_score, south_asia_outliers_gdp]).drop_duplicates()
middle_east_outliers = pd.concat([middle_east_outliers_score, middle_east_outliers_gdp]).drop_duplicates()

print("South Asia Outliers:")
print(south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])
print("\nMiddle East Outliers:")
print(middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])
```
